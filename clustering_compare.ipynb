```python

import numpy as np
import simple_wta
import ahuja
import spectral_clustering
import kmeans_clustering
import regret_matching
from time import perf_counter
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from sklearn.cluster import KMeans
n_w,n_t = (100,100)
n_c = 10
rng = np.random.RandomState(seed=1)
probs = [simple_wta.random_wta_factory(n_w,n_t) for i in range(100)]
```

```python
rng = np.random.RandomState(seed=1)
x_spectral = []
t_spectral = []
spectral_data = []
f_spectral = []
x_kmeans = []
t_kmeans = []
kmeans_data = []
f_kmeans = []

for p in probs:
    spectral_data.append(spectral_clustering.reduce_problem(p,n_c,rng=rng))
    x_spectral.append([ahuja.optimize(p,maxiters=500) for p in spectral_data[-1][0]])
    f_spectral.append(np.sum([spectral_data[-1][0][i].max_objective(x_spectral[-1][i]) for i in range(n_c)]))
    kmeans_data.append(kmeans_clustering.reduce_problem(p,n_c,rng=rng))
    x_kmeans.append([ahuja.optimize(p,maxiters=500) for p in kmeans_data[-1][0]])
    f_kmeans.append(np.sum([kmeans_data[-1][0][i].max_objective(x_kmeans[-1][i]) for i in range(n_c)]))
```

```python
plt.plot(np.array(f_spectral)-np.array(f_kmeans))
np.average(np.array(f_spectral)-np.array(f_kmeans))
```
